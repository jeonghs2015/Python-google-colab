
# 많이 하는 질문
```iloc vs loc 뭐가 다른건지??```
- iloc 는 인덱스와 컬럼을 리스트 배열로 선택하는것!
- loc 는 인덱스와 컬럼을 문자로 선택하는 것!
- 따라서 상황에 맞게 선택하는 것이 중요!

````python
# - DataFrame 행, 열, 삭제
````

````python
df = pd.DataFrame(np.arange(12).reshape(3,4), columns=['A','B','C','D'], index=['a','b','c'])
df
````

````python
df.drop(['B','C'],axis=1)
````

````python
df.drop(['b','c'],axis=0)
````

````python
df.loc[1,'C'] = '육'
````

````python
df.iloc[1,2] = 6
````

````python
df = df.drop([1],axis=0)
````

````python
df
````

````python
df[df['C']<=7]
````

````python
df[(df['B']>1) & (df['B']<9)]
````

````python
df[(df['B']>5) | (df['B']<5)]
````

````python
df[df['B'].apply(lambda x : x in [1,2,3,4,5,6])]
````

````python
df[df['C'].apply(lambda x : x in [1,6.0,7,8,9])]
````
